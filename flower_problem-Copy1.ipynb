```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#each point is length widht type(0,1)
data = [[3,   1.5, 1],
        [2,   1 ,  0],
        [4,   1.5, 1],
        [3,   1 ,  0],
        [3.5, 0.5, 1],
        [2,   0.5, 0],
        [5.5, 1,   1],
        [1,   1,   0]
       ] 

mystry_flower = [4.5,1]
```

```python
#network

#   flower type
#    0     w1,w2,length,width
#  /  \


#0     0
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_p(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
#graph making through this function
T=np.linspace(-20,20,100)
plt.plot(T,sigmoid(T),c='r')
plt.plot(T,sigmoid_p(T),c='b')
```

```python
#scatter function
plt.axis([0,6,0,6])
plt.grid()
for i in range(len(data)):
    points=data[i]
    color="r"
    if points[2]==0:
        color="b"
    plt.scatter(points[0],points[1],c=color)
```

```python
#traing the bot yepppie!!!!!..
w1=np.random.randn()
w2=np.random.randn()
b=np.random.randn()
learning_rate=0.5

costs=[]
for i in range(1,50000):
    ri=np.random.randint(len(data))
    points=data[ri]
    
    
    z=points[0]*w1+points[1]*w2+b
    pred=sigmoid(z)
    
    
    
    
    target=points[2]
    cost=np.square(pred-target)
    
    
    
    dcost_pred=2*(pred-target)
    dpred_dz = sigmoid_p(z)
    dz_dw1=points[0]
    dz_dw2=points[1]
    dz_b=1
    
    
    
    dcost_dw1=dcost_pred*dpred_dz*dz_dw1
    dcost_dw2=dcost_pred*dpred_dz*dz_dw2
    dcost_b=dcost_pred*dpred_dz*dz_b
    
    w1=w1-learning_rate*dcost_dw1
    w2=w2-learning_rate*dcost_dw2
    b=b-learning_rate*dcost_b
    if i%100==0:  
        cost_sum=0
        for j in range(len(data)):
            point=data[ri]
            z=point[0]*w1+point[1]*w2+b
            pred=sigmoid(z)
            target=point[2]
            cost_sum+=np.square(pred-target)
        costs.append(cost_sum/len(data))
    
plt.plot(costs)    
```
